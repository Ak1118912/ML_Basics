```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("train.csv")
print(data.head())
```

```python
X = data[['x']]
y = data['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
plt.scatter(data['x'], data['y'], color='blue')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter Plot of x vs y")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (Coefficient): {slope}")
print(f"Intercept: {intercept}")
```

```python
X_test_clean = X_test.dropna()
y_test_clean = y_test.dropna()
common_index = X_test_clean.index.intersection(y_test_clean.index)
X_test_clean = X_test_clean.loc[common_index]
y_test_clean = y_test_clean.loc[common_index]
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test_clean)
mse = mean_squared_error(y_test_clean, y_pred)
r2 = r2_score(y_test_clean, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R²): {r2}")
```

```python
# Plot regression line with data points
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Line")
plt.legend()
plt.show()
```
